```python
import re
pattern = "(?P<file_name>(?P<base_name>[^_]*_(?P<datetime>(?P<date>\d*)_(?P<time>\d*)))\.pkl$)"
reg = re.compile(pattern)

matches = []
for file in sorted(files):
    match = reg.search(file)
    if match is not None:
        matches.append(match)
        
for match in matches:
    print(match.groupdict())
    
    
```

```python
image_dataframe = pd.DataFrame(columns=matches[0].groupdict().keys())

for match in matches:
    image_dataframe = image_dataframe.append(match.groupdict(), ignore_index=True)

image_dataframe
#len(image_dataframe["base_name"])
```

# Unpickle Images


```python



row_dict_list = []

# create loop over dataframe to unpickle files
for basename in image_dataframe.base_name:
    
# create a new dataframe and append the unpickled information 
    erg = joblib.load(basename + ".pkl")
    
    for obj in erg:
        if isinstance(obj, tuple):
            if len(obj) == 3:
                row_dict = {}
                row_dict["image"] = basename
                row_dict["object"] = obj[0].decode("utf-8", "ignore")
                row_dict["prob"] = obj[1]
                row_dict["coord0"] = obj[2][0]
                row_dict["coord1"] = obj[2][1]
                row_dict["coord2"] = obj[2][2]
                row_dict["coord3"] = obj[2][3]
        
        row_dict_list.append(row_dict)
        
print("total objects are:",len( row_dict_list))
print("-------------------------------------------------------------")

#print(row_dict_list[100:110])

print()
for i in range(len(row_dict_list)):
    print(i, row_dict_list[i])
```

```python
object_df = pd.DataFrame(columns = row_dict_list[0].keys())
object_df
```

```python
# create new df, append all detected objects
object_df = object_df.append(row_dict_list, ignore_index=True)
```

```python
object_df
#objects_dataframe[["image", "object"]]
```

```python
# merge the two Dataframes

final_df =pd.merge(
    image_dataframe,
    object_df,
    how = "inner",          # inner join
    left_on = "base_name", # Similar keys -> foreign keys
    right_on = "image"
)

final_df
```

```python
timestamps = sorted(final_df[final_df.datetime >= "20181130_000000"].timestamp.unique())
timestamps = sorted(final_df.timestamp.unique())


print(timestamps)
unique_objects = sorted([str(x) for x in final_df["object"].unique()])
```

```python
root_df = pd.DataFrame(columns=["root_timestamp","root_object"])
import itertools
root_dict_list =[]
for element in itertools.product(timestamps, unique_objects):
    root_dict = {}   
    root_dict["root_timestamp"] = element[0]
    root_dict["root_object"] = element[1]
    root_dict_list.append(root_dict)
    
root_dict_list

root_df = root_df.append(root_dict_list)
root_df
```

```python
final_df = pd.merge(
    root_df,
    final_df,
    how = "left",
    left_on = ["root_timestamp","root_object"],
    right_on = ["timestamp", "object"]
)
```

```python
# Try to resolve these NaN values - Is this normal, or do I drop these?
final_df
```

```python
# Does not fill in all other columns with the relevant values. 

final_df.groupby(["root_timestamp", "root_object"]).count()
```

```python
# group by root (timestamp and object),  on object, count the occurances, rename object to count

sum_df = final_df.groupby(["root_timestamp","root_object"])[["object"]].count().rename({"object": "count"}, axis = 1)
sum_df
```

```python
# Get all objects in sum df

sum_df.index.get_level_values(1)
```

```python
# Build a new dataframe to perform same function as the one above.
# Print total counts of objects detected per root timestamps (Image)

sum_df_new = sum_df.reset_index().pivot(index= "root_timestamp", columns = "root_object", values = "count")
sum_df_new
```

```python
# Pickle (save) sum_df_new to use in the dashboard

sum_df_new.to_pickle("./sum_df_new.pkl")
```
